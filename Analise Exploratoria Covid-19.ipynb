```python
# Data Visualization
# Visualização de Dados

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Converting "Total Cases" to a numeric column
# Convertendo "Total Cases" para uma coluna numérica
base['Total Cases'] = pd.to_numeric(base['Total Cases'], errors='coerce')
```

```python
# Eliminating Rows whose "Total Deaths" values are empty.
# Eliminando Linhas cujos valores "Total Deaths" estejam vazios.
base.dropna(subset=['Total Deaths'], inplace=True)
```

```python
# Sorting the database by 'Total Deaths' and selecting the Top 10 countries in this regard.
# Classificando a base por 'Total Deaths' e selecionando os 10 Principais paises nesse quesito.
top_countries_deaths = base.sort_values(by='Total Deaths', ascending=False).head(10)
```

```python
#Creating a Pie chart to show the distribution of COVID-19 deaths among the main countries.
#Criando um gráfico de Pizza para mostrar a distribuição das mortes por COVID-19 entre os principais paises.

# Delete the first row (headers) of the DataFrame
# Excluir a primeira linha (cabeçalhos) do DataFrame
top_countries_deaths = top_countries_deaths.iloc[1:]

# Remove commas from values in the 'Total Deaths' column and convert to float
# Remover vírgulas dos valores na coluna 'Total Deaths' e converter para float
top_countries_deaths['Total Deaths'] = top_countries_deaths['Total Deaths'].str.replace(',', '').astype(float)


# Plot the pie chart
# Plotar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(top_countries_deaths['Total Deaths'].astype(float), labels=top_countries_deaths['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries in Total Deaths from Covid-19')
plt.axis('equal')  
plt.show()
```
