```python
# Data Visualization
# Visualização de Dados

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Convert 'Total Cases' to numeric
# Converte 'Total de Casos' para numérico
base['Total Cases'] = pd.to_numeric(base['Total Cases'], errors='coerce')
```

```python
# Drop rows with missing 'Total Cases' values
# Eliminar linhas com valores de 'Total de casos' ausentes
base.dropna(subset=['Total Cases'], inplace=True)
```

```python
# Sort the dataframe by 'Total Cases' and select top 10 countries
# Classifique o dataframe por 'Total de casos' e selecione os 10 principais países
top_countries = base.sort_values(by='Total Cases', ascending=False).head(10)
```

```python
# Create a bar plot of total cases by country
# Crie um gráfico de barras do total de casos por país
plt.figure(figsize=(12, 6))
sns.barplot(x='Total Cases', y='Country', data=top_countries)
plt.title('Top 10 Countries by Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.show()
```

```python
# Convert 'Total Deaths' to numeric
# Converte 'Total de Mortes' para numérico
base['Total Deaths'] = pd.to_numeric(base['Total Deaths'], errors='coerce')
```

```python
# Drop rows with missing 'Total Deaths' values
# Eliminar linhas com valores de 'Total de Mortes' ausentes
base.dropna(subset=['Total Deaths'], inplace=True)
```

```python
# Sort the dataframe by 'Total Deaths' and select top 10 countries
# Classifique o dataframe por 'Total Deaths' e selecione os 10 principais países
top_countries_deaths = base.sort_values(by='Total Deaths', ascending=False).head(10)
```

```python
# Create a pie chart to show the distribution of COVID-19 deaths among the top countries
# Crie um gráfico de pizza para mostrar a distribuição das mortes por COVID-19 entre os principais países
plt.figure(figsize=(8, 8))
plt.pie(top_countries_deaths['Total Deaths'], labels=top_countries_deaths['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of COVID-19 Deaths Among Top 10 Countries')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
